```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
from statsmodels.stats.multitest import multipletests
```

```python
df = pd.read_csv("170524_Atg18variants.csv")
df
```

```python
# calculate SE / cells (%)
df_number = df.groupby(["plasmid", "number"], as_index=False).sum()
df_number["SE"] = df_number["dots"] * 100 / df_number["cells"]
df_number
```

```python
# calculate mean and SEM of different plasmids
SE_mean = df_number.groupby("plasmid").mean()["SE"]
SE_sem = df_number.groupby("plasmid").std()["SE"].values / np.sqrt(3)
print(SE_mean)
print(SE_sem)
```

## Welth's t-test collected with Holm-sidak

```python
WT_SE = df_number[df_number["plasmid"] == 1007]["SE"]
atg18_SE = df_number[df_number["plasmid"] == 130]["SE"]
T56E_SE = df_number[df_number["plasmid"] == 1022]["SE"]
SLoop_SE = df_number[df_number["plasmid"] == 1025]["SE"]
FGGG_SE = df_number[df_number["plasmid"] == 1027]["SE"]
```

```python
statistic, atg18_pvalue = stats.ttest_ind(WT_SE, atg18_SE, equal_var=False)
atg18_pvalue
```

```python
statistic, T56E_pvalue = stats.ttest_ind(WT_SE, T56E_SE, equal_var=False)
T56E_pvalue
```

```python
statistic, SLoop_pvalue = stats.ttest_ind(WT_SE, SLoop_SE, equal_var=False)
SLoop_pvalue
```

```python
statistic, FGGG_pvalue = stats.ttest_ind(WT_SE, FGGG_SE, equal_var=False)
FGGG_pvalue
```

```python
multipletests(pvals=[atg18_pvalue, T56E_pvalue, SLoop_pvalue, FGGG_pvalue], alpha=0.05, method="holm-sidak")
```

## figure

```python
fontsize=15
y = -5

fig, ax = plt.subplots(figsize = (5, 2), dpi=100)
sns.swarmplot(x="plasmid", y="SE", data=df_number, size=6, color="black")
ax.errorbar(x=[0, 1, 2, 3, 4], y=SE_mean, yerr=SE_sem, capsize=5, fmt='_', ecolor='k', ms=20, mfc='None', mec='k')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.text(0, y, "none", ha="center", fontsize=15)
ax.text(1, y, "WT", ha="center", fontsize=15)
ax.text(2, y, "T56E", ha="center", fontsize=15)
ax.text(3, y, "SLoop", ha="center", fontsize=15)
ax.text(4, y, "FGGG", ha="center", fontsize=15)

ax.plot([0, 0.95], [21.5, 21.5], color='k', lw=1)
ax.text(0.3, 22.5, f'p = {atg18_pvalue:.3f}$^*$', ha='center', fontsize=fontsize-2)
ax.plot([1.05, 2], [21.5, 21.5], color='k', lw=1)
ax.text(2, 22.5, f'p = {T56E_pvalue:.3f}$^*$', ha='center', fontsize=fontsize-2)
ax.plot([1.0, 3], [25.5, 25.5], color='k', lw=1, clip_on=False)
ax.text(3, 26.5, f'p = {SLoop_pvalue:.3f}$^*$', ha='center', fontsize=fontsize-2)
ax.plot([1.0, 4], [29.5, 29.5], color='k', lw=1, clip_on=False)
ax.text(4, 30.5, f'p = {FGGG_pvalue:.3f}$^*$', ha='center', fontsize=fontsize-2)

ax.set_xlim(-0.5, 4.5)
ax.set_ylim(0, 25)
ax.set_xticklabels("")
ax.set_xlabel("")
ax.set_yticks([0, 10, 20])
ax.set_yticklabels(["0", "10", "20"], fontsize=15)
ax.set_ylabel("% of cells\n with puncta", fontsize=15, loc="center")
fig.savefig("output.jpg", bbox_inches="tight")
```
