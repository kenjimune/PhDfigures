```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
from statsmodels.stats.multitest import multipletests
```

```python
df = pd.read_csv("yRL1380_vacuole.csv")
print(df.columns)
df
```

```python
df["vacuoles*cells"] = df["how many"] * df["count"]
df
```

```python
df.info()
```

```python
def vacuole_class(number):
    if number <= 2:
        return "1-2"
    if 2 < number <= 4:
        return "3-4"
    else:
        return "5<"
```

```python
df["vacuole_class"] = df["how many"].apply(vacuole_class)
df
```

### calculate SEM

```python
df_sem = df.groupby(["strain", "vacuole_class", "number"], as_index=False).sum()
df_sem
```

```python
strains = df_sem["strain"].unique().tolist()
cells_number_dict = {}
for strain in strains:
    cells_number_dict[strain] = {}
# for index, row in df_sem.iterrows():
#     print(row["plasmid"])
```

```python
cells_number = df_sem.groupby(["strain", "number"], as_index=False).sum()
```

```python
cells_number
```

```python
for _, row in cells_number.iterrows():
    cells_number_dict[row["strain"]][row["number"]] = row["count"]
cells_number_dict
```

```python
df_sem["percentage"] = df_sem.apply(lambda row: row["count"] * 100 / cells_number_dict[row["strain"]][row["number"]], axis=1)
```

```python
df_sem
```

```python
vacuoles_percentage_mean = df_sem.groupby(["strain", "vacuole_class"], as_index=False).mean()[["strain", "vacuole_class", "percentage"]]
```

```python
vacuoles_percentage_mean
```

```python
vacuoles_percentage_sem = df_sem.groupby(["strain", "vacuole_class"], as_index=False).std()
vacuoles_percentage_sem["sem"] = vacuoles_percentage_sem["percentage"] / np.sqrt(3)
vacuoles_percentage_sem
```

```python
vacuoles_percentage_mean["sem"] = vacuoles_percentage_sem["sem"]
vacuoles_percentage_mean
```

```python
df_bar_795 = vacuoles_percentage_mean[vacuoles_percentage_mean["strain"]==795]
df_bar_897 = vacuoles_percentage_mean[vacuoles_percentage_mean["strain"]==897]
df_bar_1380 = vacuoles_percentage_mean[vacuoles_percentage_mean["strain"]==1380]
```

```python
df_bar_795
```

```python
df_bar_897
```

```python
df_bar_1380
```

```python
x = np.arange(len(df_bar_795))
distance = 0.25
width = 0.2
fig, ax = plt.subplots(figsize = (2, 2), dpi=100)
ax.bar(x=x, height="percentage", yerr="sem", data=df_bar_795, width=width, capsize=3, ecolor='black')
ax.bar(x=x+distance, height="percentage", yerr="sem", data=df_bar_897, width=width, capsize=3, ecolor='black')
ax.bar(x=x+distance*2, height="percentage", yerr="sem", data=df_bar_1380, width=width, capsize=3, ecolor='black')


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.set_xticklabels("")
ax.set_xlabel("")
ax.set_xticks(x+distance)
ax.set_xticklabels(["1-2", "3-4", "5<"], fontsize=12)
ax.set_xlabel("Number of vacuoles per cell", fontsize=13, loc="center")

ax.set_yticks([0, 20, 40, 60, 80])
ax.set_yticklabels(["0", "20", "40", "60", "80"], fontsize=12)
ax.set_ylabel("% of total", fontsize=13, loc="center")

fig.savefig("yRL1380_vacuole_class.jpg", bbox_inches="tight")
```

### vacuole average

```python
df
```

```python
df_number = df.groupby(["strain", "number"], as_index=False).sum()
df_number
```

```python
df_number["vacuole/cell"] = df_number["vacuoles*cells"] / df_number["count"]
df_number
```

```python
vacuoles_mean = df_number.groupby(["strain"], as_index=False).mean()["vacuole/cell"].values
vacuoles_sem = df_number.groupby(["strain"], as_index=False).std()["vacuole/cell"].values / np.sqrt(3)
print(vacuoles_mean, vacuoles_sem)
```

## Welch's t-test

```python
wt_vacuoles = df_number[df_number["strain"] == 795]["vacuole/cell"]
atg18_vacuoles = df_number[df_number["strain"] == 897]["vacuole/cell"]
yRL1380_vacuoles = df_number[df_number["strain"] == 1380]["vacuole/cell"]
print(wt_vacuoles)
print(atg18_vacuoles)
print(yRL1380_vacuoles)
```

```python
statistic, wt_pvalue = stats.ttest_ind(atg18_vacuoles, wt_vacuoles, equal_var=False)
wt_pvalue
```

```python
statistic, yRL1380_pvalue = stats.ttest_ind(atg18_vacuoles, yRL1380_vacuoles, equal_var=False)
yRL1380_pvalue
```

```python
multipletests(pvals=[wt_pvalue, yRL1380_pvalue], alpha=0.05, method="holm-sidak")
```
