```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multitest import multipletests
```

```python
df = pd.read_csv("yRL910_old.csv")
df
```

```python
# calculate SE / cells (%)
df_number = df.groupby(["aTc_conc", "number", "time"], as_index=False).sum()
df_number["SE"] = df_number["dots"] * 100 / df_number["cells"]
df_number
```

```python
# filter different aTc concentration
df_0 = df_number[df_number["aTc_conc"]==0]
df_1 = df_number[df_number["aTc_conc"]==1]
df_10 = df_number[df_number["aTc_conc"]==10]
df_0
```

```python
# calculate average of SE number in each aTc concentration
df_0_mean = df_0.groupby('time', as_index=False).SE.mean()
df_1_mean = df_1.groupby('time', as_index=False).SE.mean()
df_10_mean = df_10.groupby('time', as_index=False).SE.mean()

# calculate sem of SE number in each aTc concentration
df_0_sem = df_0.groupby('time').SE.std().values / np.sqrt(3)
df_1_sem = df_1.groupby('time').SE.std().values / np.sqrt(3)
df_10_sem = df_10.groupby('time').SE.std().values / np.sqrt(3)

# calculate average of SE number and extract values in each aTc concentration
df_0_mean_values = df_0_mean['SE'].values
df_1_mean_values = df_1_mean['SE'].values
df_10_mean_values = df_10_mean['SE'].values
```

## data visualization

```python
marker_map = {
    1: 'o',  # ä¸¸
    2: 's',  # å››è§’
    3: '^'   # ä¸‰è§’
}
x_diff = {
    1: -1,
    2: 0,
    3: 1
}
fontsize=15
fig, ax = plt.subplots(figsize = (3, 2), dpi=100)
# aTc = 0 ng/ml
ax.plot("time", "SE", data=df_0_mean, color="black", label="0 ng/ml")
for timepoint in df_0['time'].unique():
    subset = df_0[df_0['time'] == timepoint]
    for _, row in subset.iterrows():
        x = timepoint + x_diff[row['number']]
        y = row['SE']
        marker = marker_map[row['number']]
        ax.scatter(x, y, s=12, marker=marker, color='black')
ax.errorbar(x=list(range(0, 150, 30)), y=df_0_mean_values, yerr=df_0_sem, 
            capsize=5, fmt='none', ecolor='k', ms=20, mfc='None', mec='k')
ax.text(75, 0, 'No aTc', color='k', fontsize=fontsize-2)

# aTc = 1 ng/ml
color = '#ff7f0e'
ax.plot("time", "SE", data=df_1_mean, color=color, label="1 ng/ml")
for timepoint in df_1['time'].unique():
    subset = df_1[df_1['time'] == timepoint]
    for _, row in subset.iterrows():
        x = timepoint + x_diff[row['number']]
        y = row['SE']
        marker = marker_map[row['number']]
        ax.scatter(x, y, s=12, marker=marker, color=color)
ax.errorbar(x=list(range(0, 150, 30)), y=df_1_mean_values, yerr=df_1_sem, 
            capsize=5, fmt='none', ecolor=color, ms=20, mfc='None', mec='k')
ax.text(75, 5, '1 ng/ml', color=color, fontsize=fontsize-2)

# aTc = 10 ng/ml
color = '#00008B'
ax.plot("time", "SE", data=df_10_mean, color=color, label="10 ng/ml")
for timepoint in df_10['time'].unique():
    subset = df_10[df_10['time'] == timepoint]
    for _, row in subset.iterrows():
        x = timepoint + x_diff[row['number']]
        y = row['SE']
        marker = marker_map[row['number']]
        ax.scatter(x, y, s=12, marker=marker, color=color)
ax.errorbar(x=list(range(0, 150, 30)), y=df_10_mean_values, yerr=df_10_sem, 
            capsize=5, fmt='none', ecolor=color, ms=20, mfc='None', mec='k')   
ax.text(30, 10, '10 ng/ml', color=color, fontsize=fontsize-2)

# p value
ax.text(130, 10, f'p = {pvalue10:.3f}$^*$', fontsize=fontsize-2, color='#00008B')
ax.text(130, 7.5, f'p = {pvalue1:.3f}$^*$', fontsize=fontsize-2, color='#ff7f0e')
ax.text(130, 5, 'at 120 min', fontsize=fontsize-2, color='k')
ax.plot([128, 128], [2, 13], color='k', lw=1, clip_on=False)

# figure settings
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# x and y axis settings
ax.set_xlim(-5, 125)
ax.set_ylim(-1, 16)
ax.set_xlabel("Time (min)", fontsize=fontsize)
ax.set_xticks([0, 30, 60, 90, 120])
ax.set_xticklabels(["0", "30", "60", "90", "120"], fontsize=fontsize)
ax.set_ylabel("% of cells\n with puncta", fontsize=fontsize)
ax.set_yticks([0, 2.5, 5.0, 7.5, 10.0, 12.5, 15.0])
ax.set_yticklabels(["0", "2.5", "5", "7.5", "10", "12.5", "15"], fontsize=fontsize)
# ax.legend()
fig.savefig("output_ver2.jpg", bbox_inches="tight")
```

## paired t-test at 120 min corrected by Holm-Sidak

```python
# extract t=120 min
df_120min = df[df['time']==120].groupby(['aTc_conc', 'number'], as_index=False).sum()
df_120min["SE"] = df_120min["dots"] * 100 / df_120min["cells"]
df_120min
```

```python
# extract SE number in each aTc concentration
aTc_0 = df_120min[df_120min['aTc_conc']==0]['SE']
aTc_1 = df_120min[df_120min['aTc_conc']==1]['SE']
aTc_10 = df_120min[df_120min['aTc_conc']==10]['SE']
aTc_10
```

```python
# calculate p-value
# this experiment is done in 3 different days, so used paired t-test.
statistic, pvalue1 = stats.ttest_rel(aTc_0, aTc_1)
statistic, pvalue10 = stats.ttest_rel(aTc_0, aTc_10)
print(f"aTc 0 vs 1 ng/ml pvalue: {pvalue1:.5f}")
print(f"aTc 0 vs 10 ng/ml pvalue: {pvalue10:.5f}")
# multiple testing correction
multipletests(pvals=[pvalue1, pvalue10], alpha=0.05, method="holm-sidak")
```
