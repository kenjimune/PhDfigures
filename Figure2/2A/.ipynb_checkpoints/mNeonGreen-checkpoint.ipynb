```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("fpbase_spectra.csv")
df
```

```python
fig, ax = plt.subplots(figsize = (10, 3), dpi=100)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.plot('wavelength', 'mNeonGreen ex', data=df, color='black')
ax.plot('wavelength', 'mNeonGreen em', data=df, color='#00008B')

ax.axvline(x=488, color='#00B8CF', linestyle='--', linewidth=2)
ax.axvline(x=514, color='#00D300', linestyle='--', linewidth=2)
ax.axvline(x=561, color='#CCFF00', linestyle='--', linewidth=2)
ax.axvline(x=594, color='#FF5E00', linestyle='--', linewidth=2)

ax.text(430, 0.2, 'Excitation', fontsize=14)
ax.text(565, 0.2, 'Emission', fontsize=14, color='#00008B')

ax.text(488, -0.16, '488', ha='center', fontsize=12, color='#00B8CF')
ax.text(514, -0.16, '514', ha='center', fontsize=12, color='#00D300')
ax.text(561, -0.16, '561', ha='center', fontsize=12, color='#CCFF00')
ax.text(594, -0.16, '594', ha='center', fontsize=12, color='#FF5E00')

ax.set_xlabel("Wavelength (nm)", fontsize=12)
ax.set_ylabel("Ratio to the maximum value", fontsize=12, loc="center")

ax.set_xlim(400, 599)

fig.savefig("mNeonGreen.jpg", bbox_inches="tight")
plt.show()
```
