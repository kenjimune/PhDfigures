```python
import pandas as pd
```

```python
df = pd.read_csv("Kenji16.csv")
df.head()
```

```python
df['SE'] = df['SE'].fillna(0)
df.head()
```

```python
df.columns
```

```python
df.set_index(' ', inplace=True)
```

```python
df.head()
```

```python
def ref_peri(number):
    if number == '-':
        return number
    else:
        return df['Perim.'].iloc[int(number)-1]
```

```python
df['ref_peri'] = df['reference'].apply(ref_peri)
```

```python
def md_ratio(row):
    if (row['classification']=='-') | (row['classification']=='un'):
        return '-'
    elif row['classification']=='bud':
        return row['Perim.'] / float(row['ref_peri'])
    else:
        return float(row['ref_peri']) / row['Perim.']
```

```python
df['ratio'] = df.apply(md_ratio, axis=1)
```

```python
df
```

```python
df.info()
```

```python
def classification(row):
    if row['classification']=='-':
        return '-'
    elif row['classification']=='un':
        return 'no_bud'
    elif (row['classification']=='bud') & (float(row['ratio'])<=0.5):
        return 'small_daughter'
    elif (row['classification']=='bud') & (float(row['ratio'])>0.5):
        return 'large_daughter'
    elif (row['classification']=='mother') & (float(row['ratio'])<=0.5):
        return 'small_mother'
    else:
        return 'large_mother'
```

```python
df['five_class'] = df.apply(classification, axis=1)
```
