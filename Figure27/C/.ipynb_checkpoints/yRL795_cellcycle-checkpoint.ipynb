```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from matplotlib.backends.backend_pdf import PdfPages

df = pd.read_csv("cell_cycle.csv")
df
```

```python
df_date = df.groupby(["date", "class"], as_index=False).sum()
df_date
```

```python
df_date["SE_cell"] = df_date["number.1"] * 100 / df_date["total"]
```

```python
df_date
```

```python
dates = df_date["date"].unique().tolist()
```

```python
date_dict = {}
for date in dates:
    date_dict[date] = df_date[(df_date["date"]==date) & (df_date["class"]=="mother")]["SE_cell"].values[0]
print(date_dict)
```

```python
date_dict
```

```python
for idx, row in df_date.iterrows():
    
```

```python
df_1 = df_date[df_date["date"]==30524]
df_1["SE_cell_mother"] = df_1["SE_cell"] / date_dict[30524]
df_1
```

```python
df_2 = df_date[df_date["date"]==190924]
df_2["SE_cell_mother"] = df_2["SE_cell"] / date_dict[190924]
df_2
```

```python
df_3 = df_date[df_date["date"]==300424]
df_3["SE_cell_mother"] = df_3["SE_cell"] / date_dict[300424]
df_3
```

```python
# ç¸¦
df_conc = pd.concat([df_1, df_2, df_3], axis=0)
```

```python
df_conc
```

```python
cycle_mean = df_conc.groupby("class").mean()["SE_cell_mother"].values
cycle_sem = df_conc.groupby("class").std()["SE_cell_mother"].values / np.sqrt(3)
```

```python
print(cycle_mean)
print(cycle_sem)
```

```python
fig, ax = plt.subplots(figsize = (3, 2), dpi=100)
sns.swarmplot(x="class", y="SE_cell_mother", hue = "date", palette="deep", data=df_conc, size=6,  color="black")
ax.errorbar(x=list(range(3)), y=cycle_mean, yerr=cycle_sem, capsize=5, fmt='_', ecolor='k', ms=20, mfc='None', mec='k')

ax.legend().remove()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_ylim(0.5, 1.5)
y = 0.3

ax.text(0, y, "daughter\ncells", ha="center", va="center",fontsize=12)
ax.text(1, y, "mother\ncells", ha="center", va="center", fontsize=12)
ax.text(2, y, "unbudded", ha="center", va="center", fontsize=12)
# ax.text(1, 0, "Cell stage", ha="center", va="center", fontsize=12)

ax.set_xticklabels("")
ax.set_xlabel("")

ax.set_ylabel("SE per cells\n(relative to mother cells)", fontsize=12, loc="center")

ax.set_yticks([0.6, 0.8, 1.0, 1.2, 1.4])
ax.set_yticklabels(["0.6", "0.8", "1.0", "1.2", "1.4"], fontsize=10)

fig.savefig("yRL795_cell_cycle.jpg", bbox_inches="tight")
```
