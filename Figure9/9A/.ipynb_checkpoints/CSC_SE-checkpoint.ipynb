```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
from statsmodels.stats.multitest import multipletests
```

```python
df = pd.read_csv("CSC_mutant_3.csv")
df
```

```python
# calculate SE / cells (%)
df_number = df.groupby(["strain", "number"], as_index=False).sum()
df_number["SE"] = df_number["dots"] * 100 / df_number["cells"]
df_number
```

```python
# calculate average and sem in different strains
# wt, vps26, vps29, vps35
SE_mean = df_number.groupby("strain").mean()["SE"]
SE_sem = df_number.groupby("strain").std()["SE"].values / np.sqrt(3)
print(SE_mean)
print(SE_sem)
```

## Welth's t-test collected with Holm-sidak

```python
WT_SE = df_number[df_number["strain"] == 795]["SE"]
vps26_SE = df_number[df_number["strain"] == 1063]["SE"]
vps29_SE = df_number[df_number["strain"] == 1066]["SE"]
vps35_SE = df_number[df_number["strain"] == 1069]["SE"]
```

```python
statistic, vps26_pvalue = stats.ttest_ind(WT_SE, vps26_SE, equal_var=False)
vps26_pvalue
```

```python
statistic, vps29_pvalue = stats.ttest_ind(WT_SE, vps29_SE, equal_var=False)
vps29_pvalue
```

```python
statistic, vps35_pvalue = stats.ttest_ind(WT_SE, vps35_SE, equal_var=False)
vps35_pvalue
```

```python
multipletests(pvals=[vps26_pvalue, vps29_pvalue, vps35_pvalue], alpha=0.05, method="holm-sidak")
```

## figure

```python
y = -7
fontsize = 15
fig, ax = plt.subplots(figsize = (4, 2), dpi=100)
sns.swarmplot(x="strain", y="SE", data=df_number.iloc[:12], size=6, color="black")
ax.errorbar(x=[0, 1, 2, 3], y=SE_mean[0:4], yerr=SE_sem[0:4], capsize=5, fmt='_', ecolor='k', ms=20, mfc='None', mec='k')
# label of each strains
ax.text(0, y, "WT", ha="center", fontsize=fontsize)
ax.text(1, y, "vps26Δ", ha="center", fontstyle="italic", fontsize=fontsize)
ax.text(2, y, "vps29Δ", ha="center", fontstyle="italic", fontsize=fontsize)
ax.text(3, y, "vps35Δ", ha="center", fontstyle="italic", fontsize=fontsize)

ax.plot([0, 1], [25, 25], color='k', lw=1)
ax.text(0.5, 26, f'p = {vps26_pvalue:.3f}$^*$', ha='center', fontsize=fontsize-2)
ax.plot([0, 2], [30, 30], color='k', lw=1)
ax.text(1.5, 31, f'p = {vps29_pvalue:.3f}$^*$', ha='center', fontsize=fontsize-2)
ax.plot([0, 3], [35, 35], color='k', lw=1)
ax.text(2.5, 36, f'p = {vps35_pvalue:.3f}$^*$', ha='center', fontsize=fontsize-2)

# frame
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# x and y axis
ax.set_xticklabels("")
ax.set_xlabel("")
ax.set_xlim(-0.5, 3.5)
ax.set_ylim(0, 35)
ax.set_yticks([0, 10, 20, 30])
ax.set_yticklabels(["0", "10", "20", "30"], fontsize=fontsize)
ax.set_ylabel("Cells with foci (%)", fontsize=fontsize, loc="center")

fig.savefig("output.jpg", bbox_inches="tight")
```
